```python
import re

text = "Some text with Gegevens    2020 and Gegevens 2019, and more text"
match = re.search(r'Gegevens\s*(\d{4})', text)

if match:
    year = match.group(1)
    print(year)  # This will print the year of the first match
else:
    print("No match found")
```

```python
import re

text = "Totale Bezoldiging some text 5.000, Bezoldiging more text 134.500, yet more text 12.345"
n = 2  # Number of matches you want to retrieve

# Find the part of the text that follows "Totale Bezoldiging"
pattern = r"(Totale Bezoldiging|Totaal Bezoldiging|Bezoldiging)(.*)"
match = re.search(pattern, text)
if match:
    following_text = match.group(2)

    # Find all number patterns in the following text
    number_pattern = r'\d{1,3}\.\d+'
    matches = re.findall(number_pattern, following_text)

    # Get the first 'n' matches
    first_n_matches = matches[:n]
    print(first_n_matches)
else:
    print("No matches found")

```
